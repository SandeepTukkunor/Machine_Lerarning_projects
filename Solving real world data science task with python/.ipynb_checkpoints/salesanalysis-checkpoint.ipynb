But as we have the data of 12 months, we have to read all the data sets at a time, so we have to make use of list dir

### For reading all the data sets, we have to make use of listdir 

```python
#import listdir
from os import listdir
```

```python
files = [file for file in os.listdir("./Sales_Data")] # this is reading all the datasets 

all_months_data = pd.DataFrame() #take empty df and concat 


for file in files:
    df = pd.read_csv("./Sales_Data/" + file) # to read all the data sets 
    all_months_data = pd.concat([all_months_data,df]) #combine all the data sets into one 
    
```

```python
all_months_data.head()
```

```python
all_months_data.to_csv("all_data.csv", index=False) # we have made a new csv file 
```

### Task 3. Add a sales column 

```python
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
```

###  Q1. What was the best month for sales?And how much was earned that month

```python
# we can just group by 
results = all_data.groupby('Month').sum()
results
```

We ca see that dec was the month with most sale, we can plot for more visualisation

```python
# we can plot the sales results 
months = range(1,13)
plt.bar(months,results["Sales"])
plt.xticks(months) # to see all the months labled 
plt.ylabel("sales in US$")
plt.xlabel("Months")
```
